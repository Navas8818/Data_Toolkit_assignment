 6.Create a 3D NumPy array of shape (3,3,3) containing random integers between 1 and 10 Perform the
  following operations:

 a) Find the indices of the maximum values along each depth level (third axis).

 b) Perform element wise multiplication of between both array

```python
import numpy as np
random = np.random.randint(1,10,size=(3,3,3))
print("3D array is ", random)
```

```python
#Find the indices of the maximum values along each depth level (third axis)
max_indices = np.argmax(random, axis= 2)
print("max_indices is " , max_indices)
```

```python
#Perform element wise multiplication of between both array
import numpy as np
random = np.random.randint(1,10,size=(3,3,3))
max_indices = np.argmax(random, axis= 2)
element_wisemultiplication = np.multiply(random, max_indices)
print("Element_wise", element_wisemultiplication)
```
